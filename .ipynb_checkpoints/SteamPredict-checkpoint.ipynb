### 数据介绍

**赛题背景**

火力发电的基本原理是：燃料在燃烧时加热水生成蒸汽，蒸汽压力推动汽轮机旋转，然后汽轮机带动发电机旋转，产生电能。在这一系列的能量转化中，影响发电效率的核心是锅炉的燃烧效率，即燃料燃烧加热水产生高温高压蒸汽。锅炉的燃烧效率的影响因素很多，包括锅炉的可调参数，如燃烧给量，一二次风，引风，返料风，给水水量；以及锅炉的工况，比如锅炉床温、床压，炉膛温度、压力，过热器的温度等。

**赛题描述**

经脱敏后的锅炉传感器采集的数据（采集频率是分钟级别），根据锅炉的工况，预测产生的蒸汽量。<br>
网址：[天池原始数据][1]
[1]:https://tianchi.aliyun.com/competition/introduction.htm?spm=5176.11409106.5678.1.537cd6fe3rXmFh&raceId=231693/   "天池原始数据"

**数据说明**

数据分成训练数据（train.txt）和测试数据（test.txt），其中字段”V0”-“V37”，这38个字段是作为特征变量，”target”作为目标变量。选手利用训练数据训练出模型，预测测试数据的目标变量，排名结果依据预测结果的MSE（mean square error）。

**结果提交**

选手需要提交测试数据的预测结果（txt格式，只有1列预测结果）。

**结果评估**

预测结果以mean square error作为评判标准。

### 模型训练与特征选择

#### 简单线性回归

我们选择使用一个简单的线性回归模型来对单一特征进行训练。

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def simple_linear_model(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, input_feature):
    reg = LinearRegression()
    reg.fit(zhengqi_train_data.as_matrix(columns=[input_feature]), zhengqi_train_y)
    MSE = mean_squared_error(
        zhengqi_val_y, reg.predict(zhengqi_val_data.as_matrix(columns=[input_feature]))
    )
    return MSE  # 均方误差
```

```python
np.random.seed(21)
MSE = simple_linear_model(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, 'V0')
print('Validation MSE for V0 is : %s' %MSE)
```

**mean_squared_error:均方误差**

同样的，我们可以对数据中的所有特征进行相同的训练，使用上面的简单线性回归模型来评估哪个特征更能拟合模型。

```python
estimate_result = pd.DataFrame(columns=['feature', 'Val_MSE'])

np.random.seed(21)

for feature in zhengqi_train_data.columns:
    MSE = simple_linear_model(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, feature)
    estimate_result = estimate_result.append({'feature': feature, 'Val_MSE': MSE}, ignore_index=True)
```

```python
estimate_result.sort_values(by='Val_MSE').head(10)
```

#### 多元线性回归

接下来再简单的修改线性回归函数，在上面的基础上将多个特征作为输入用来训练和拟合蒸汽量数据。

```python
def multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, input_feature):
    reg = LinearRegression()
    reg.fit(zhengqi_train_data.as_matrix(columns=input_feature), zhengqi_train_y)
    MSE = mean_squared_error(
        zhengqi_val_y, reg.predict(zhengqi_val_data.as_matrix(columns=input_feature))
    )
    return MSE
```

```python
# 根据预测数据预测结果target
def mutiple_regression_result(zhengqi_train_data, zhengqi_train_y, zhengqi_test_data, input_feature):
    reg = LinearRegression()
    reg.fit(zhengqi_train_data.as_matrix(columns=input_feature), zhengqi_train_y)
    zhengqi_test_y = reg.predict(zhengqi_test_data.as_matrix(columns=input_feature))
    return zhengqi_test_y
```

现在尝试几种不同的组合

```python
feature_combinations = [
    ['V0', 'V8', 'V27'],
    ['V0', 'V8', 'V37'],
    ['V1', 'V27', 'V8'],
    ['V31', 'V8', 'V37']
]

np.random.seed(21)

for combination in feature_combinations:
    MSE = multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, combination)
    print('MSE for %s is %s' %(','.join(combination), MSE))
```

对比三种特征组合，可以看出选择V0、V8、V37进行组合作为特征，训练的效果更好。

我们还可以选择拟合效果较好的特征用于构建多项式特征，比如选择V0、V1构建一个二次项特征。

```python
zhengqi_train_data['V0_squared'] = zhengqi_train_data['V0'].apply(lambda x: x**2)
zhengqi_val_data['V0_squared'] = zhengqi_val_data['V0'].apply(lambda x: x**2)
zhengqi_train_data['V1_squared'] = zhengqi_train_data['V1'].apply(lambda x: x**2)
zhengqi_val_data['V1_squared'] = zhengqi_val_data['V1'].apply(lambda x: x**2)
zhengqi_test['V0_squared'] = zhengqi_test['V0'].apply(lambda x: x**2)
zhengqi_test['V1_squared'] = zhengqi_test['V1'].apply(lambda x: x**2)

squared_combinations = [
    ['V0', 'V0_squared'],
    ['V1', 'V1_squared']
]

np.random.seed(21)

for squared_combination in squared_combinations:
    MSE = multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, squared_combination)
    print('MSE for %s is %s' %(','.join(squared_combination), MSE))
```

#### 特征选择

通过选择不同的组合和构造多项式特征可以获得性能更好的模型，但是仍然存在许多问题，我们并不知道该选择哪些特征来进行组合或者构造多项式特征能够在验证集上得到更小的MSE，通常如果将所有特征都加入到模型训练中很容易导致过拟合，所以我们需要进行特征提取。这里我们采用随机森林（randomforest）来做特征提取。

我们先使用所有特征来训练模型查看效果。

```python
np.random.seed(21)

MSE = multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, zhengqi_train_data.columns)
print('MSE for all feature before construction is %s' %MSE)
```

然后我们再构造其他多项式特征使得构造的特征能够使模型达到过拟合

```python
zhengqi_train_data['V0_cubed'] = zhengqi_train_data['V0'].apply(lambda x: x**3)
zhengqi_val_data['V0_cubed'] = zhengqi_val_data['V0'].apply(lambda x: x**3)
zhengqi_test['V0_cubed'] = zhengqi_test['V0'].apply(lambda x: x**3)

zhengqi_train_data['V27_squared'] = zhengqi_train_data['V27'].apply(lambda x: x**2)
zhengqi_val_data['V27_squared'] = zhengqi_val_data['V27'].apply(lambda x: x**2)
zhengqi_test['V27_squared'] = zhengqi_test['V27'].apply(lambda x: x**2)

zhengqi_train_data['V8'] = zhengqi_train_data['V27'] * zhengqi_train_data['V8']
zhengqi_val_data['V8'] = zhengqi_val_data['V27'] * zhengqi_val_data['V8']
zhengqi_test['V8'] = zhengqi_test['V27'] * zhengqi_test['V8']
```

使用随机森林模型来做特征提取

```python
zhengqi_train_data.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor

np.random.seed(21)

rf_reg = RandomForestRegressor(n_estimators=50, verbose=1)
rf_reg.fit(zhengqi_train_data, zhengqi_train_y)

combine_lists = lambda item: [item[0], item[1]]
feature_importance = list(map(combine_lists, zip(zhengqi_train_data.columns, rf_reg.feature_importances_)))
feature_importance = pd.DataFrame(
    feature_importance, columns=['feature', 'importance']
).sort_values(by='importance', ascending=False)
```

**sklearn.ensemble.RandomForestRegressor()**

参数：<br>
- n_estimators: 森林里树的数量，通常数量越大，效果越好，但是计算时间也会随之增加。<br>
此外要注意，当树的数量超过一个临界值之后，算法的效果并不会很显著地变好。<br>
- max_features: 分割节点时考虑的特征的随机子集的大小。这个值越低，方差减小的越多，<br>但是偏差的增大也越多。根据经验，回归问题中使用max_features = n_features，分类问题是用max_features = sqrt(n_feature(其中n_features是特征的个数) 是比较好的默认值。<br>
- verbose:(default=0)是否显示任务进程。<br>
- max_depth = None 和 min_samples_split = 2 结合通常会有不错的效果（即生成完全的树）。

```python
feature_importance.head(10)
```

过滤重要性小于0.001的所有特征包含已构造的特征在内

```python
filter_feature = feature_importance[feature_importance['importance'] > 0.001]['feature'].tolist()
```

```python
np.random.seed(21)

MSE = multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, filter_feature)
print('MSE for filtered features after construction is %s' % MSE)
MSE = multiple_regression(zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, zhengqi_train_data.columns)
print('MSE for all features after construction is %s' %MSE)

# 预测测试集中的target值
target_predict = mutiple_regression_result(zhengqi_train_data, zhengqi_train_y, zhengqi_test, filter_feature)
```

#### 尝试其他模型

上述训练中我们都是采用的简单的线性回归模型，还有很多其他的回归模型我们可以选择，比如岭回归模型、随机森林回归模型等等可以尝试，通常可以得到不同的结果。

```python
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

np.random.seed(21)

def other_regression_model(reg, zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, input_feature):
    reg.fit(zhengqi_train_data.as_matrix(columns=input_feature), zhengqi_train_y)
    MSE = mean_squared_error(
        zhengqi_val_y, reg.predict(zhengqi_val_data.as_matrix(columns=input_feature))
    )
    return reg, MSE

def other_regression_model_test(reg, zhengqi_train_data, zhengqi_train_y, zhengqi_test, input_feature):
    reg.fit(zhengqi_train_data.as_matrix(columns=input_feature), zhengqi_train_y)      
    zhengqi_test_y = reg.predict(zhengqi_test.as_matrix(columns=input_feature))
    return zhengqi_test_y

ridge = Ridge()
dt_gre = DecisionTreeRegressor()
rf_reg = RandomForestRegressor()
ridge, MSE = other_regression_model(ridge, zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, filter_feature)
print('MSE for ridge_regression is %s' % MSE)
dt_gre, MSE = other_regression_model(dt_gre, zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, filter_feature)
print('MSE for DecisionTreeRegressor is %s' % MSE)
rf_reg, MSE = other_regression_model(rf_reg, zhengqi_train_data, zhengqi_train_y, zhengqi_val_data, zhengqi_val_y, filter_feature)
print('MSE for RandomForestRegressor is %s' % MSE)

zhengqi_test_result = other_regression_model_test(ridge, zhengqi_train_data, zhengqi_train_y, zhengqi_test, filter_feature)
```
