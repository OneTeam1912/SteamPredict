```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score   # 交叉验证
params = [1, 2, 3, 4, 5, 6]
test_scores = []   # 用来存放不同参数模型的结果值
for param in params:
    clf = XGBRegressor(max_depth = param)
    test_score = np.sqrt(-cross_val_score(clf, zhengqi_train, zhengqi_train_y, cv=10, scoring = 'neg_mean_squared_error')) #cv：选择每次测试折数  平均方差(Mean squared error)：评价指标是准确度,可以不要
    test_scores.append(np.mean(test_score))
plt.plot(params, test_scores)
plt.title('max_depth vs CV Error')
plt.show()
    
#     xgb = XGBRegressor(max_depth = 5)
#     xgb.fit(zhengqi_train_data, zhengqi_train_y)
#     y_final = np.expm1(xgb.predict(zhengqi_test))
```
